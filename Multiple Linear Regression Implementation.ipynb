### Multiple Linear Regression Implementation

Our mode Yhat = Xw, where X is an N by D matrix, and w is a D by 1 vector. We have the same error function as the simple linear regression, but the expression for the prediction part is changed. We use partial derivative solving for w, the weights, and the w = np.inv(X.T.dot(X)).dot(X.T.dot(Y)). But we are not going to use inverse, we are going to use np.linalg.solve() to find w.

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
# load the data
# We need to have a bias term. it's implicit in the solution.
X = []
Y = []
for line in open('data_2d.csv'):
    x1, x2, y = line.strip().split(',')
    X.append([float(x1), float(x2), 1])
    Y.append(float(y))
```

```python
# turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# plot the data to see what the data looks like
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y);
```

```python
# Calculate weights, which are resulted from our hand calculation
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
```

```python
# Yhat = Xw, note the order here. Each sample of X is a row.
# Since X is an N by D matrix, w of D by 1, so w is put behind
Yhat = X.dot(w)
```

```python
# Calculate R-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1)/d2.dot(d2)
```
